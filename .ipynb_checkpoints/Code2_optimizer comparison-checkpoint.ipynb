### 几中不同优化算法的比较

```python
# 载入必要的函数库
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from keras import Model
from keras.layers import Dense,Flatten,Input
from keras.optimizers import SGD,RMSprop,Adam

from keras.utils import to_categorical 
```

```python
## 载入mnist数据集
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data() 

## 转换为one - hot型向量
Y_train=to_categorical(y_train)
Y_test=to_categorical(y_test)

print(Y_train.shape)
print(Y_train[0])
```

```python
'''
 实验：构建 Multi-layer Nueral Network 模型 
 超参数：input_size   输入尺寸 MNIST为 [28,28]
         hidden_op    隐藏层节点数 
         output_size  输出尺寸 MNSIT为 10分类
'''

input_size = [28,28]
hidden_op = 100
output_size = 10

## STEP.1 建立模型结构
def build_model(input_size,hidden_op,output_size):
                                                   
    input_layer = Input(input_size)            
    x = input_layer                              
    x = Flatten()(input_layer)                  
    x = Dense(hidden_op,activation = 'relu')(x)       
    x = Dense(output_size,activation = 'softmax')(x)
    output_layer=x
    model=Model(input_layer,output_layer)
    return model
```

```python
sgd = SGD(lr = 0.001)
momentum = SGD(lr = 0.001, momentum = 0.9, nesterov = False)
rmsp = RMSprop(lr = 0.001)
adam = Adam(lr = 0.001)

# 创建一个优化器字典，用于后面循环
dict = {'SGD': sgd, 'Momentum': momentum, 'RMSprop': rmsp, 'Adam': adam}
```

```python
train_loss = []           
for key in dict.keys():
    
    print("Now we are using " + key + " optimizer !")
    model = build_model(input_size,hidden_op,output_size)
    model.compile(loss='categorical_crossentropy',optimizer=dict[key],metrics=['accuracy'])
    history = model.fit(X_train,Y_train, validation_data=(X_test,Y_test), batch_size=1000, epochs=50)
    train_loss.append(history.history['loss'])
```

```python
epochs = np.arange(1,51,1)
plt.plot(epochs,train_loss[0],label = "SGD")
plt.plot(epochs,train_loss[1],label = "Momentum")
plt.plot(epochs,train_loss[2],label = "RMSprop")
plt.plot(epochs,train_loss[3],label = "Adam")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.ylim(0,1)
plt.legend()
plt.show()
```
